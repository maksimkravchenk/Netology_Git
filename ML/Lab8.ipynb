```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
house_price = load_boston()
X = pd.DataFrame(house_price.data, columns=house_price.feature_names)
y = pd.DataFrame(house_price.target, columns=['MEDV'])
df = pd.concat([X,y], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
y_train.info()
```

```python
X_test.info()
```

```python
y_test.info()
```

```python
depth = [1,2,3,4,5,6,7]

dt_regs = {}

for i in depth:
    dt_regs[f'depth={i}'] = DecisionTreeRegressor(max_depth = i).fit(X_train, y_train)
    dt_regs[f'depth={i}'].score(X_train, y_train)
    dt_regs[f'depth={i}'].score(X_test, y_test)
    print(f"Train Model depth={i} score on is equal:", dt_regs[f'depth={i}'].score(X_train, y_train))
    print(f"Test Model depth={i} score is equal:", dt_regs[f'depth={i}'].score(X_test, y_test))
```
