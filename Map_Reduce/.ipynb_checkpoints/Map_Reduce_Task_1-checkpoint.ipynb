```python
##Reducer

previous_word = None
word_count = 0

with open('keywords_sorted.csv', 'r', encoding='utf-8') as f:
    for line in f:
        word, one = line.strip().split(',')
        one = int(one)
        if previous_word:
            if previous_word == word: 
                word_count += one
            else: 
                print(previous_word, word_count)
                previous_word = word
                word_count = one
        else:
            previous_word = word
            word_count = one
    print(previous_word, word_count)
f.close()
```
